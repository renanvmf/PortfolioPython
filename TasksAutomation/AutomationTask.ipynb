```python
import pyautogui
import pyperclip
import time
import tkinter
import pandas as pd

pyautogui.pause = 2

# Acessando Google Drive
pyautogui.hotkey("ctrl", "t")
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")
pyautogui.hotkey("ctrl","v")
#pyautogui.press("backspace")
#pyautogui.press("home")
#pyautogui.press("delete")
pyautogui.press("enter")
time.sleep(10)

#Acessando a pasta e fazendo download do arquivo
pyautogui.click(x=430, y=300 , clicks = 2)
time.sleep(2)
pyautogui.click(x=491, y=347)
time.sleep(1)
pyautogui.click(x=1159, y=180)
time.sleep(1)
pyautogui.click(x=961, y=586)

#Lendo o arquivo baixado para pegar os indicadores

df = pd.read_excel(r"/home/renanvmf/Downloads/Vendas - Dez.xlsx")
display(df)

#Calculando os indicadores

faturamento = df["Valor Final"].sum()
qtde_produtos = df["Quantidade"].sum()

#Enviando e-mail com os valores calculados

pyautogui.hotkey("ctrl","t")
pyautogui.write("mail.google.com")
pyautogui.press("delete")
pyautogui.press("enter")
time.sleep(5)
pyautogui.click(x=166, y=198)
time.sleep(2)
pyautogui.write("renanvmf@hotmail.com")
time.sleep(1)
pyautogui.press("tab")
pyautogui.press("tab")
time.sleep(1)
pyautogui.write("Indicadores de vendas")
pyautogui.press("tab")
time.sleep(1)
texto = f"""
Prezados,

Segue os valores dos indicadores de vendas:

Faturamento = R${faturamento:,.2f}
Quantidade de produtos = {qtde_produtos:,}

Atenciosamente


"""
pyperclip.copy(texto)
pyautogui.hotkey("ctrl" , "v")
time.sleep(3)
pyautogui.press("enter")
```
